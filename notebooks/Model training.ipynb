```python
import pandas as pd
```

```python
df = pd.read_csv("data/gemstone.csv")
```

```python
df.head()
```

```python
df = df.drop(labels = 'id', axis =1)
```

```python
#Independent and dependent features
X = df.drop(labels='price',axis=1)
y = df[['price']]
```

```python
X.head()
```

```python
#Define which column should be ordinal encoded and which should be scaled
numerical_col = df.select_dtypes(exclude='object').columns
categorical_col = df.select_dtypes(include='object').columns
```

```python
numerical_col = numerical_col.drop(labels='price')
```

```python
numerical_col
```

```python
#Define custom ranking for each ordinal variables
cut_categories = ['Fair','Good','Very Good','Premium','Ideal']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
color_categories= ['D','E','F','G','H','I','J']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
## Numerical Pipeline
num_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
)

##Categorical pipeline
cat_pipeline = Pipeline(
    steps = [
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinal_encoder',OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
        ('scaler',StandardScaler())
    ]
)

preprocessor = ColumnTransformer([
        ('num_pipeline',num_pipeline,numerical_col),
        ('cat_pipeline',cat_pipeline,categorical_col)
])
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state =42)
```

```python
X.columns
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z'])
X_test = pd.DataFrame(preprocessor.transform(X_test),columns=['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z'])
```

```python
X_train
```

```python
X_test.head()
```

```python
#Model Training
```

```python
from sklearn.linear_model import LinearRegression,Lasso,ElasticNet,Ridge
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#Train multiple models
```

```python
import numpy as np
```

```python
models = {
    'Linear_regression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'ElasticNet':ElasticNet()
}

model_perf = []
for i,j in models.items():
    #print(j)
    model = j
    #Model training
    model.fit(X_train,y_train)

    #Model prediction
    y_pred = model.predict(X_test)

    r2score = r2_score(y_test,y_pred)
    mse = mean_squared_error(y_test,y_pred)
    mae = np.sqrt(mse)

    model_perf.append({
        'Model Name': i,
        'R_squared': r2score,
        'mse':mse,
        'mae':mae
    })


```
